### An:Neの比率を満たすような組みを探しましょう

```python
import numpy as np                     # import and create an alias for numpy for numerical operations
import pandas as pd                    # "  pandas to handle spreadsheets

import matplotlib as mpl               # "  matplotlib, the fundamental plotting toolset
from matplotlib import pyplot as plt   # "  pyplot, a very commonly used subset of tools.
import seaborn as sns   
```

```python
df_coefficients_2 = pd.read_csv('coefficients_of_every_a_and_b_2.csv')
df_coefficients_2
```

### 適切なL範囲条件

```python
stable_set_df_2 = df_coefficients_2[(df_coefficients_2['v1_Ne'] * df_coefficients_2['v1_Ab'] > 0 ) & (df_coefficients_2['v2_Ne']>0) & (df_coefficients_2['v2_Ab'] < 0)]
```

### $Ab:Ne=7:3$

```python
Ab_Ne_ratio_set_df = df_coefficients_2[abs(df_coefficients_2['v1_Ab'] / (df_coefficients_2['v1_Ne'] + df_coefficients_2['v2_Ne']) - 7/3) < 10**(-2)]
```

### その周辺

```python
Ab_Ne_ratio_set_df_1 = df_coefficients_2[abs(df_coefficients_2['v1_Ab'] / (df_coefficients_2['v1_Ne'] + df_coefficients_2['v2_Ne']) - 6.1/3) < 10**(-3)]
```

```python
Ab_Ne_ratio_set_df_2 = df_coefficients_2[abs(df_coefficients_2['v1_Ab'] / (df_coefficients_2['v1_Ne'] + df_coefficients_2['v2_Ne']) - 6.5/3) < 10**(-2)]
```

```python
Ab_Ne_ratio_set_df_3 = df_coefficients_2[abs(df_coefficients_2['v1_Ab'] / (df_coefficients_2['v1_Ne'] + df_coefficients_2['v2_Ne']) - 9/3) < 10**(-1.5)]
```

```python
Ab_Ne_ratio_set_df_4 = df_coefficients_2[abs(df_coefficients_2['v1_Ab'] / (df_coefficients_2['v1_Ne'] + df_coefficients_2['v2_Ne']) - 11/3) < 10**(-1.5)]
```

```python
# 条件を満たすa, bの値だけを新しいDataFrameに格納
stable_set_df = stable_set_df_2[['LSiSi/LNaNa', 'LSiSi/LAlAl']]
Ab_Ne_ratio_df = Ab_Ne_ratio_set_df[['LSiSi/LNaNa', 'LSiSi/LAlAl']]
Ab_Ne_ratio_df_1 = Ab_Ne_ratio_set_df_1[['LSiSi/LNaNa', 'LSiSi/LAlAl']]
Ab_Ne_ratio_df_2 = Ab_Ne_ratio_set_df_2[['LSiSi/LNaNa', 'LSiSi/LAlAl']]
Ab_Ne_ratio_df_3 = Ab_Ne_ratio_set_df_3[['LSiSi/LNaNa', 'LSiSi/LAlAl']]
Ab_Ne_ratio_df_4 = Ab_Ne_ratio_set_df_4[['LSiSi/LNaNa', 'LSiSi/LAlAl']]
# insert a log version of area and stream power for later
stable_set_df.insert(1,  "log_(LSiSi/LNaNa)", np.log10(stable_set_df["LSiSi/LNaNa"].values))
stable_set_df.insert(3,  "log_(LSiSi/LAlAl)", np.log10(stable_set_df["LSiSi/LAlAl"].values))

Ab_Ne_ratio_df.insert(1,  "log_(LSiSi/LNaNa)", np.log10(Ab_Ne_ratio_df["LSiSi/LNaNa"].values))
Ab_Ne_ratio_df.insert(3,  "log_(LSiSi/LAlAl)", np.log10(Ab_Ne_ratio_df["LSiSi/LAlAl"].values))

Ab_Ne_ratio_df_1.insert(1,  "log_(LSiSi/LNaNa)", np.log10(Ab_Ne_ratio_df_1["LSiSi/LNaNa"].values))
Ab_Ne_ratio_df_1.insert(3,  "log_(LSiSi/LAlAl)", np.log10(Ab_Ne_ratio_df_1["LSiSi/LAlAl"].values))

Ab_Ne_ratio_df_2.insert(1,  "log_(LSiSi/LNaNa)", np.log10(Ab_Ne_ratio_df_2["LSiSi/LNaNa"].values))
Ab_Ne_ratio_df_2.insert(3,  "log_(LSiSi/LAlAl)", np.log10(Ab_Ne_ratio_df_2["LSiSi/LAlAl"].values))

Ab_Ne_ratio_df_3.insert(1,  "log_(LSiSi/LNaNa)", np.log10(Ab_Ne_ratio_df_3["LSiSi/LNaNa"].values))
Ab_Ne_ratio_df_3.insert(3,  "log_(LSiSi/LAlAl)", np.log10(Ab_Ne_ratio_df_3["LSiSi/LAlAl"].values))

Ab_Ne_ratio_df_4.insert(1,  "log_(LSiSi/LNaNa)", np.log10(Ab_Ne_ratio_df_4["LSiSi/LNaNa"].values))
Ab_Ne_ratio_df_4.insert(3,  "log_(LSiSi/LAlAl)", np.log10(Ab_Ne_ratio_df_4["LSiSi/LAlAl"].values))

sns.set_theme()   # set the seaborn basic theme as a default (light grey background with white grid)

# create a figure and define the axes within it, set two axes side-by-side
fig, ax1 = plt.subplots(figsize=(8, 6),ncols=1)

# using seaborn to make a graph
sns.scatterplot(x='log_(LSiSi/LNaNa)',y='log_(LSiSi/LAlAl)',data=stable_set_df,color='blue',linewidth=0,s=19,ax=ax1,zorder=1)
sns.scatterplot(x='log_(LSiSi/LNaNa)',y='log_(LSiSi/LAlAl)',data=Ab_Ne_ratio_df,color='yellow',linewidth=0,s=19,ax=ax1,zorder=2)
sns.scatterplot(x='log_(LSiSi/LNaNa)',y='log_(LSiSi/LAlAl)',data=Ab_Ne_ratio_df_1,color='red',linewidth=0,s=19,ax=ax1,zorder=3)
sns.scatterplot(x='log_(LSiSi/LNaNa)',y='log_(LSiSi/LAlAl)',data=Ab_Ne_ratio_df_2,color='black',linewidth=0,s=19,ax=ax1,zorder=4)
sns.scatterplot(x='log_(LSiSi/LNaNa)',y='log_(LSiSi/LAlAl)',data=Ab_Ne_ratio_df_3,color='green',linewidth=0,s=19,ax=ax1,zorder=5)
sns.scatterplot(x='log_(LSiSi/LNaNa)',y='log_(LSiSi/LAlAl)',data=Ab_Ne_ratio_df_4,color='white',linewidth=0,s=19,ax=ax1,zorder=6)
# `s` sets the size of the dots
# `linewidth` sets the thickenss of a the outline of each dot. Set to zero when the points are overlapping closely

ax1.axis('equal')
plt.tight_layout() # important to pack everything within the frame
fig.savefig('Onsagers_coefficients2.pdf')
```
