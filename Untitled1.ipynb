```python
## HSVからBGRを経由して白黒に変更する ##
img_bgr = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)
img_bw  = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
```
